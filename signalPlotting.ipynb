```python
import numpy as np
import pickle as pkl
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import raw fluorescence
signals = np.load(recording + 'signals_0.pkl')
signalTraces = signals['ROIs']['raw'][0]
# signalTraces.shape
signalsDf = pd.DataFrame(signalTraces)
signalDF = signalsDf.T

#import dfof
dfof = np.load(recording +'dFoF_0.pkl')
dfofTraces = dfof['ROIs']['traces']
a,b,c = dfofTraces.shape
new_dfofTraces = dfofTraces.reshape([a,b])
dfofTracesDf = pd.DataFrame(new_dfofTraces)
dFoFDfTraces = dfofTracesDf.T

#import neuropil from Fnpli.npy
npil = np.load(recording + 's2p/suite2p/plane0/Fneu.npy')
npilDF = pd.DataFrame(npil)
npilT = npilDF.T

#import neuropil from signals_0.pkl
npilTraces = signals['ROIs']['npil'][0]
npilTracesDf = pd.DataFrame(npilTraces)
npilDf2 = npilTracesDf.T

# plot stuff
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(25,6))
ax1.plot(signalDF) #= signalDF.plot(legend=False)
ax1.set_title('raw Fluorescence from signal_0.pkl')
ax2.plot(dFoFDfTraces) #dFoFDfTraces.plot(legend=False)
ax2.set_title('dfof from dFoF_0.pkl')
ax3.plot(npilT)
ax3.set_title('npil from Fneu.npy')
ax4.plot(npilDf2)
ax4.set_title('npil from signals_0.pkl')
fig.suptitle(recording)
```
